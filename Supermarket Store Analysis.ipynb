# Data Visualization

```python
df.head()
```

```python
# Sort the dataframe by 'Store_Sales' and take the top 5 areas
top_5_df = df.sort_values('Store_Sales', ascending=False).head(5)

# Bar chart (Top 5 Store Area by Store Sales)
top_5_df.plot.bar(x = 'Store_Area', y = 'Store_Sales', figsize=(10,6))

# label the axis
plt.xlabel('Store_Area')
plt.ylabel('Store_Sales')

# set the title
plt.title('Top 5 Store Area Vs Store Sales')

# show the visual
plt.show()
```

```python
# Sort the dataframe by 'Items_Available' and take the top 5
top_5_df = df.sort_values('Items_Available', ascending=False).head(5)

# Creating a horizontal bar chart to visualize Items Available, Daily Customer Count and Store Sales for top 5
top_5_df[['Items_Available', 'Daily_Customer_Count', 'Store_Sales']].plot.barh(figsize=(10, 6))

# set the title
plt.title('Top 5 Items Available, Daily Customer Count, and Store Sales')

# show the visual
plt.show()
```

```python
# Calculate correlations between variables
correlations = df.corr()
print(correlations)
```

```python
# Create a new figure for the heatmap with a size of 10x6
plt.figure(figsize=(10, 6))

# Generate a heatmap using seaborn to visualize the correlations between variables. 
# The 'annot' parameter is set to True to write the data value in each cell. 
# The colormap 'hot' is used to map normalized data values to colors.
sns.heatmap(correlations, annot=True, cmap='hot')

# Set the title of the heatmap
plt.title('Correlation Heatmap')

# Display the heatmap
plt.show()
```

```python
# Sort the dataframe by 'Store_Sales' and take the top 5
top_5_df = df.sort_values('Store_Sales', ascending=False).head(5)

# Plot a Scatter plot for Store Sales by Items Available and Store Area for top 5
plt.scatter(top_5_df['Store_Area'], top_5_df['Items_Available'], c=top_5_df['Store_Sales'])

# Add labels and title
plt.xlabel('Store Area')
plt.ylabel('Items Available')
plt.colorbar(label='Store Sales')

# Show Plot
plt.show()
```

```python
# Sort the dataframe by 'Items_Available' and take the top 10
top_10_df = df.sort_values('Items_Available', ascending=False).head(10)

# Plot scatter plot for top 10
top_10_df.plot.scatter('Items_Available', 'Store_Area', figsize=(10,6))

# Add labels and title
plt.xlabel('Items Available')
plt.ylabel('Store Area')
plt.title('Top 10 Store Area Vs Items Available')

# Show plot
plt.show()
```
